### Шаг 1. Моделирование реального АБ-теста с зависимостями целевой метрики

То есть определяем кол-во пользователем, случайно разбиваем на 2 группы. Далее моделируем метрику в предпериоде (пусть это будет выручка). Создаем набор дополнительных ковариат. Дальше моделируем целевую метрику как некоторую функциональную зависимость метрик предпериода с добавлением шума. И в конце к тестовой группе добавляем некое воздействие, которое мы и будем оценивать.

```python
# Фиксируем случайное зерно для воспроизводимости
np.random.seed(42)

# Количество пользователей
n_users = 5000

# Симуляция user_id
user_ids = np.arange(n_users)

# Разбиение на контроль и тест (50/50)
groups = np.random.choice(["control", "test"], size=n_users)

# Симуляция предпериодной метрики (например, выручка за предыдущий период)
metric_pre = np.random.normal(loc=100, scale=20, size=n_users)

# Дополнительные ковариаты, которые объясняют дисперсию метрики
covariate_1 = np.random.normal(loc=50, scale=15, size=n_users)  # влияет на метрику
covariate_2 = np.random.uniform(30, 70, size=n_users)  # влияет на метрику

# Задаём истинную зависимость целевой метрики
true_treatment_effect = 5  # Реальный эффект в тестовой группе

# Формируем метрику с зависимостью от предпериода и ковариат
metric = (
    0.6 * metric_pre + 0.3 * covariate_1 + 0.2 * covariate_2 +  # Влияние признаков
    np.random.normal(loc=0, scale=10, size=n_users)  # Шум
)

# Добавляем эффект только для тестовой группы
metric += (groups == "test") * true_treatment_effect

# Формируем DataFrame
df = pd.DataFrame({
    "user_id": user_ids,
    "group": groups,
    "metric_pre": metric_pre,
    "covariate_1": covariate_1,
    "covariate_2": covariate_2,
    "metric": metric
})
```

### Шаг 2. Расчет эффектов с использованием различных алгоритмов

- Классический критерий t-test Стьюдента

```python
### 1. Базовый A/B-анализ без корректировки
control = df[df["group"] == "control"]["metric"]
test = df[df["group"] == "test"]["metric"]

results_abs = absolute_ttest(control, test)

print(f"Истинный эффект составляет {true_treatment_effect}")

print("\n*** Результаты A/B теста ***\n")

# Абсолютный эффект
print("Абсолютный эффект:")
print(f"P-value: {results_abs.pvalue:.2e}")
print(f"Эффект: {results_abs.effect:.2f} " 
      f"(ДИ: [{results_abs.left_bound:.2f}; {results_abs.right_bound:.2f}])\n")


print(f"Дисперсия оценки среднего составляет {results_abs.var:.2f}")
```

Видим, что истинный эффект попадает в доверительный интервал, то есть он оценен корректно, нет ошибки второго рода. При этом критерий верно отвергает H0, но доверительный интервал получился от 4.42 до 6.24.


- Классический CUPED

```python
control = df[df["group"] == "control"]["metric"]
test = df[df["group"] == "test"]["metric"]
control_before = df[df["group"] == "control"]["metric_pre"]
test_before = df[df["group"] == "test"]["metric_pre"]


results_cuped = cuped_ttest(control, test, control_before, test_before)

print(f"Истичнный эффект составляет {true_treatment_effect}")

print("\n*** Результаты A/B теста ***\n")

# Абсолютный эффект
print("Абсолютный эффект (классический CUPED):")
print(f"P-value: {results_cuped.pvalue:.2e}")
print(f"Эффект: {results_cuped.effect:.2f} " 
      f"(ДИ: [{results_cuped.left_bound:.2f}; {results_cuped.right_bound:.2f}])\n")
print(f"Дисперсия оценки среднего составляет {results_cuped.var:.2f}")

```

Видим, что истинный эффект попадает в доверительный интервал, то есть он оценен корректно, нет ошибки второго рода. При этом критерий верно отвергает H0, но доверительный интервал получился от 4.67 до 5.92. То есть мы точнее оценили наш эффект. При этом дисперсия составляет 0.10.

```python
1 - results_cuped.var / results_abs.var
```

Снижение дисперсии составляет 53%. Этим объясняется существенное сужение доверительного интервала.

- ML CUPED

```python
from sklearn.ensemble import GradientBoostingRegressor

### 3. ML-CUPED (градиентный бустинг для предсказания метрики)
X = df[["metric_pre", "covariate_1", "covariate_2"]]
y = df["metric"]

# Разделим на контроль и тест для предсказаний
X_control = X[df["group"] == "control"]
y_control = y[df["group"] == "control"]

# Обучаем модель на контрольной группе (чтобы избежать утечки информации из теста)
model = GradientBoostingRegressor()
model.fit(X_control, y_control)

# Делаем предсказания на всех данных
df["metric_pred"] = model.predict(X)

# Вычисляем ML-CUPED
theta_ml = np.cov(df["metric"], df["metric_pred"])[0, 1] / np.var(df["metric_pred"])
df["metric_ml_cuped"] = df["metric"] - theta_ml * df["metric_pred"]

# Новый t-test на ML-CUPED метрике
control_ml_cuped = df[df["group"] == "control"]["metric_ml_cuped"]
test_ml_cuped = df[df["group"] == "test"]["metric_ml_cuped"]

results_ml_cuped = absolute_ttest(control_ml_cuped, test_ml_cuped)

print(f"Истичнный эффект составляет {true_treatment_effect}")

print("\n*** Результаты A/B теста ***\n")

# Абсолютный эффект
print("Абсолютный эффект:")
print(f"P-value: {results_ml_cuped.pvalue:.2e}")
print(f"Эффект: {results_ml_cuped.effect:.2f} " 
      f"(ДИ: [{results_ml_cuped.left_bound:.2f}; {results_ml_cuped.right_bound:.2f}])\n")


print(f"Дисперсия оценки среднего составляет {results_ml_cuped.var:.2f}")
```

Видим, что истинный эффект попадает в доверительный интервал, то есть он оценен корректно, нет ошибки второго рода. При этом критерий верно отвергает H0, но доверительный интервал получился от 4.67 до 5.75. То есть мы точнее оценили наш эффект. При этом дисперсия составляет 0.08.

```python
1 - results_ml_cuped.var / results_cuped.var
```

Снижение дисперсии еще составило 24% по сравнению с классическим CUPED.

```python
1 - results_ml_cuped.var / results_abs.var
```

По сравнению с первоначальной историей снижение дисперсии составляет 64%.

Выше были точечные оценки, основанные на конкретной выборке. Теперь с помощью симуляций Монте-Карло проверим наш результат.

```python
from sklearn.linear_model import LinearRegression
```

```python
def generate_data(true_effect=2, n=1000, add_prepre=True):
    np.random.seed()
    
    # Базовые ковариаты
    covariate_1_prepre = np.random.normal(50, 10, n)
    covariate_2_prepre = np.random.uniform(0, 1, n)
    covariate_3_prepre = np.random.normal(100, 1, n)
    covariate_4_prepre = np.random.normal(30, 1, n)
    
    # Ковариаты предпериода
    covariate_1_pre = covariate_1_prepre * 1.3 + np.random.normal(0, 1.5, n)
    covariate_2_pre = covariate_2_prepre * 1.7 + np.random.normal(0, 0.5, n) 
    covariate_3_pre = covariate_3_prepre * 1.8 + np.random.normal(0, 1.5, n)
    covariate_4_pre = covariate_4_prepre * 3 + np.random.normal(0, 0.5, n)     

    covariate_1 = covariate_1_pre * 1.3 + np.random.normal(0, 1.5, n)
    covariate_2 = covariate_2_pre * 1.7 + np.random.normal(0, 0.5, n)
    covariate_3 = covariate_3_pre * 1.8 + np.random.normal(0, 1.5, n)
    covariate_4 = covariate_4_pre * 3 + np.random.normal(0, 0.5, n)    
    
    # Предпредпериод
    metric_prepre = covariate_1_prepre * 0.3 + covariate_2_prepre * 2 + np.random.normal(0, 1.5, n)
    
    # Предпериод зависит от предпредпериода
    metric_pre = metric_prepre * 1.8 + covariate_1_pre * 1.2 + covariate_2_pre * 1.5 + covariate_3_pre * 1.8 + \
                    covariate_4_prepre * 3.2 + np.random.normal(0, 1, n)
    
    # Разбиение на группы
    group = np.random.choice(['control', 'test'], n, p=[0.5, 0.5])
    effect = true_effect * (group == 'test')
    
    # Текущий период зависит от предпериода и ковариат
    metric = (
            metric_pre * 0.1
             + covariate_1 * 1.2 
             + covariate_2 * 1.5
             + covariate_3 * 1.8
             + covariate_4 * 3.2
             + effect 
             + np.random.normal(0, 5, n))
    
    data = {
        'group': group,
        'metric_prepre': metric_prepre,
        'covariate_1_pre': covariate_1_pre,
        'covariate_2_pre': covariate_2_pre,
        'covariate_3_pre': covariate_3_pre,
        'covariate_4_pre': covariate_4_pre,        
        'covariate_1': covariate_1,
        'covariate_2': covariate_2,
        'covariate_3': covariate_3, 
        'covariate_4': covariate_4,         
        'metric_pre': metric_pre,        
        'metric': metric
    }
    
    return pd.DataFrame(data)


# Функция для сохранения дисперсий
def run_simulation(n_simulations=1000, true_effect=2, n=1000, alpha=0.05):
    variances = {
        'base': [],
        'cuped': [],
        'ml_cuped': []
    }

    pvalues = {
        'base': [],
        'cuped': [],
        'ml_cuped': []        
    }
    # счетчики для подсчета ошибок, когда истинный эффект не попадает в ДИ (контроль уровня значимости альфа)
    bad_cnt_abs = 0
    bad_cnt_cuped = 0
    bad_cnt_ml_cuped = 0
    
    # Добавляем tqdm для основного цикла
    for _ in tqdm(range(n_simulations), desc="Симуляции", unit="exp"):
        df = generate_data(true_effect=true_effect, n=n)
        
        # Базовый метод
        control = df[df["group"] == "control"]["metric"]
        test = df[df["group"] == "test"]["metric"]
        results_abs = absolute_ttest(control, test)
        bad_cnt_abs += not(results_abs.left_bound < true_effect < results_abs.right_bound)        
        variances['base'].append(results_abs.var)
        pvalues['base'].append(results_abs.pvalue)
        
        # CUPED
        control_before = df[df["group"] == "control"]["metric_pre"]
        test_before = df[df["group"] == "test"]["metric_pre"]
        results_cuped = cuped_ttest(control, test, control_before, test_before)
        bad_cnt_cuped += not(results_cuped.left_bound < true_effect < results_cuped.right_bound)
        variances['cuped'].append(results_cuped.var)
        pvalues['cuped'].append(results_cuped.pvalue)        
        
        # ML-CUPED
        X = df[["metric_prepre", "covariate_1_pre", "covariate_2_pre", "covariate_3_pre", "covariate_4_pre"]]
        y = df["metric_pre"]
        
        X_control = X[df["group"] == "control"]
        y_control = y[df["group"] == "control"]
        
        # Добавляем tqdm для обучения модели (опционально)
#         model = GradientBoostingRegressor()
        model = LinearRegression()
        model.fit(X_control, y_control)
        
        df["metric_pred"] = model.predict(df[["metric_pre", "covariate_1", "covariate_2", "covariate_3", "covariate_4"]]
                                          .rename({'metric_pre':'metric_prepre',
                                                  'covariate_1':'covariate_1_pre',
                                                  'covariate_2':'covariate_2_pre',
                                                  'covariate_3':'covariate_3_pre',
                                                  'covariate_4':'covariate_4_pre',
                                                  }, axis=1))
        theta_ml = np.cov(df["metric"], df["metric_pred"])[0, 1] / np.var(df["metric_pred"])
        df["metric_ml_cuped"] = df["metric"] - theta_ml * df["metric_pred"]
        
        control_ml = df[df["group"] == "control"]["metric_ml_cuped"]
        test_ml = df[df["group"] == "test"]["metric_ml_cuped"]
        results_ml = absolute_ttest(control_ml, test_ml)
        bad_cnt_ml_cuped += not(results_ml.left_bound < true_effect < results_ml.right_bound)        
        variances['ml_cuped'].append(results_ml.var)
        pvalues['ml_cuped'].append(results_ml.pvalue)            
    
    bad_cnt = {
        'base': bad_cnt_abs,
        'cuped': bad_cnt_cuped,
        'ml_cuped': bad_cnt_ml_cuped
    }    
    
    return variances, bad_cnt, pvalues
```

```python
# Запуск симуляции
np.random.seed(42)
variances, bad_cnt, pvalues = run_simulation(1000)
```

```python
# Соберем данные в удобный формат для визуализации
variances_df = pd.DataFrame({
    "Метод": ["Базовый"] * len(variances["base"]) +
             ["CUPED"] * len(variances["cuped"]) +
             ["ML-CUPED"] * len(variances["ml_cuped"]),
    "Дисперсия": np.concatenate([variances["base"], variances["cuped"], variances["ml_cuped"]])
})

# Построим KDE-график распределений дисперсий
plt.figure(figsize=(10, 6))
sns.kdeplot(data=variances_df, x="Дисперсия", hue="Метод", fill=True, alpha=0.4)
plt.axvline(np.mean(variances["base"]), color="blue", linestyle="--", label="Среднее (Базовый)")
plt.axvline(np.mean(variances["cuped"]), color="orange", linestyle="--", label="Среднее (CUPED)")
plt.axvline(np.mean(variances["ml_cuped"]), color="green", linestyle="--", label="Среднее (ML-CUPED)")
plt.legend()
plt.title("Распределение дисперсий для различных методов корректировки")
plt.xlabel("Значение дисперсии")
plt.ylabel("Плотность вероятности")
plt.grid()
plt.show()
```
